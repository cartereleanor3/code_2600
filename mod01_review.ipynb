### What are the feature names?

```python
feature_names = Credit.columns.tolist()
print(feature_names)
```

### What are the data types of the feature values (i.e., numeric, categorical)

```python
numeric = ['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education', 'Balance']
categorical = ['Own', 'Student', 'Married', 'Region']
print(f'Numeric: {numeric}\nCategorical: {categorical}')
```

### Find the mean, median, and mode of the feature "Age."

```python
# mean
mean_age = Credit['Age'].mean()
print(mean_age)
```

```python
# median
median_age = Credit['Age'].median()
print(median_age)
```

```python
# mode
mode_age = Credit['Age'].mode()
for i in mode_age:
    print(i)
```

### Create a subset of Credit that only contains the columns Age, Region, and Income. Call it "Subset."

```python
Subset = Credit[['Age', 'Region', 'Income']]
Subset
```

### From this subset, create two partitions: 
- One that contains records for Ages below 50, and one that contains Ages 50+
- Call them "Younger" and "Older"

```python
Younger = Subset[Subset['Age'] < 50]
Younger
```

```python
Older = Subset[Subset['Age'] >= 50]
Older
```

### Use these partitions to compare the mean and median Income for each age group (less than 50 vs. 50+)

```python
# mean of younger
mean_income_younger = Younger['Income'].mean()
print(mean_income_younger)
```

```python
# mean of older
mean_income_older = Older['Income'].mean()
print(mean_income_older)
```

```python
# median of younger
median_income_younger = Younger['Income'].median()
print(median_income_younger)
```

```python
# median of older
median_income_older = Older['Income'].median()
print(median_income_older)
```

### What does the difference between mean and median suggest about the distribution of income?

The mean is larger than the median for both the older and younger subset, suggesting the distribution of data is skewed right.

### Compare the average income for students vs. non-students.

```python
Students = Credit[Credit['Student'] == 'Yes']
Non_Students = Credit[Credit['Student'] == 'No']
#mean for students
student_mean_income = Students['Income'].mean()
print(student_mean_income)
```

```python
#mean for non-students
non_student_mean_income = Non_Students['Income'].mean()
print(non_student_mean_income)
```

```python
#median for students
student_median_income = Students['Income'].median()
print(student_median_income)
```

```python
#median for non-students
non_student_median_income = Non_Students['Income'].median()
print(non_student_median_income)
```

Often, banks will use a variable called _utilization_ to assess credit risk.
Utilization is the ratio of balance to limit (i.e., what % of the credit limit is being used?)
### Add a feature to the Credit dataset to measure utilization

```python
Credit['Utilization'] = Credit['Balance'] / Credit ['Limit']
Credit
```

### What is the maximum utilization value in this set? 

```python
max_utilization = Credit['Utilization'].max()
print(max_utilization)
```
